# Train Neural Network





# Parse and prepare the data

Parse the CSV files and transforms them to a format that can be used to train the fully connected neural network.

If you've recorded additional vibrations, update the `vibration` list with the names of the additional CSV files.


```python
# Importation des bibliothèques nécessaires

import matplotlib.pyplot as plt  # Bibliothèque pour tracer des graphiques (non utilisée ici, mais utile pour l'analyse)
import numpy as np  # Bibliothèque pour les calculs numériques et manipulation des matrices
import pandas as pd  # Utilisée pour lire et traiter les fichiers CSV contenant les données
import tensorflow as tf  # Bibliothèque de machine learning pour entraîner le modèle

# Affichage de la version de TensorFlow pour vérifier l’installation

print(f"TensorFlow version = {tf.__version__}\n")

# Définition des types de vibrations à classifier

VIBRATIONS = ["infinity", "HeartBeat"]  #  Liste des classes de vibrations (deux types ici)

# Définition des paramètres liés aux enregistrements de vibrations

SAMPLES_PER_VIBRATION = 119  #  Nombre d'échantillons par vibration
NUM_VIBRATIONS = len(VIBRATIONS)  # Nombre total de types de vibrations (2 dans ce cas)

# Création d’une matrice d’encodage one-hot pour la classification

ONE_HOT_ENCODED_VIBRATIONS = np.eye(NUM_VIBRATIONS)  

# Initialisation des listes pour stocker les données d'entrée (features) et de sortie (labels)

inputs = []  #  Liste qui contiendra les signaux de vibration (accélération et gyroscope)
outputs = []  # Liste des labels one-hot correspondant aux types de vibrations

# Parcours de chaque type de vibration pour lire et traiter les données correspondantes

for vibration_index in range(NUM_VIBRATIONS):
    vibration = VIBRATIONS[vibration_index]  # Sélection du nom de la vibration en cours de traitement
    print(f"Processing index {vibration_index} for vibration '{vibration}'.")

    # Sélection de la sortie associée dans la matrice one-hot
    
    output = ONE_HOT_ENCODED_VIBRATIONS[vibration_index]

    # Chargement des données du fichier CSV correspondant à la vibration actuelle
    
    df = pd.read_csv(vibration + ".csv") 

    # Suppression des valeurs manquantes pour éviter les erreurs
    
    df = df.dropna().reset_index(drop=True)  

    # Calcul du nombre total d’enregistrements dans le fichier CSV
    
    num_recordings = int(df.shape[0] / SAMPLES_PER_VIBRATION)  
    print(f"\tThere are {num_recordings} recordings of the vibration {vibration}.")

    # Traitement des enregistrements individuels
    
    for i in range(num_recordings):
        tensor = []  # Liste temporaire pour stocker les données de vibration sous forme de tenseur

        # Extraction des données pour chaque échantillon de la vibration en cours
        
        for j in range(SAMPLES_PER_VIBRATION):
            index = i * SAMPLES_PER_VIBRATION + j  # Calcul de l’index de l’échantillon actuel

            # Normalisation des données pour les adapter au modèle (échelle entre 0 et 1)
            
            tensor += [
                (df['aX'][index] + 4) / 8,  #  Normalisation des données d'accélération (X)
                (df['aY'][index] + 4) / 8,  #  Normalisation des données d'accélération (Y)
                (df['aZ'][index] + 4) / 8,  #  Normalisation des données d'accélération (Z)
                (df['gX'][index] + 2000) / 4000,  # Normalisation des données du gyroscope (X)
                (df['gY'][index] + 2000) / 4000,  # Normalisation des données du gyroscope (Y)
                (df['gZ'][index] + 2000) / 4000   # Normalisation des données du gyroscope (Z)
            ]

        # Ajout des données prétraitées dans la liste des entrées
        
        inputs.append(tensor)

        # Ajout du label one-hot correspondant dans la liste des sorties
        
        outputs.append(output)

# Conversion des listes en tableaux numpy pour une compatibilité avec TensorFlow

inputs = np.array(inputs)  # Transformation de la liste des entrées en tableau numpy
outputs = np.array(outputs)  # Transformation de la liste des labels en tableau numpy

# Indication que le traitement des données est terminé

print("Data set parsing and preparation complete..")
```

## Randomize and split the input and output pairs for training

Randomly split input and output pairs into sets of data: 60% for training, 20% for validation, and 20% for testing.

  - the training set is used to train the model
  - the validation set is used to measure how well the model is performing during training
  - the testing set is used to test the model after training

```python
# Mélanger l'ordre des entrées pour assurer une répartition équilibrée entre l'entraînement, le test et la validation

num_inputs = len(inputs)
randomize = np.arange(num_inputs)
np.random.shuffle(randomize)

# Remplacer les index consécutifs (0, 1, 2, etc.) par les index mélangés

inputs = inputs[randomize]
outputs = outputs[randomize]

# Diviser les enregistrements (groupes d'échantillons) en trois ensembles : entraînement, test et validation

TRAIN_SPLIT = int(0.6 * num_inputs)  # 60 % des données pour l'entraînement
TEST_SPLIT = int(0.2 * num_inputs + TRAIN_SPLIT)  # 20 % des données pour le test, le reste pour la validation

inputs_train, inputs_test, inputs_validate = np.split(inputs, [TRAIN_SPLIT, TEST_SPLIT])
outputs_train, outputs_test, outputs_validate = np.split(outputs, [TRAIN_SPLIT, TEST_SPLIT])

print("Mélange et division du jeu de données terminés.")
```

## Build & Train the Model

Build and train a [TensorFlow](https://www.tensorflow.org) model using the high-level [Keras](https://www.tensorflow.org/guide/keras) API.

```python
# Construction du modèle de classification des vibrations

model = tf.keras.Sequential([  # Création d'un modèle séquentiel
    tf.keras.layers.Dense(50, activation='relu'),  # Première couche cachée avec 50 neurones et activation ReLU
    tf.keras.layers.Dense(15, activation='relu'),  # Deuxième couche cachée avec 15 neurones et activation ReLU
    tf.keras.layers.Dense(NUM_VIBRATIONS, activation='softmax')  # Couche de sortie avec activation Softmax pour classification
])

# Compilation du modèle

model.compile(
    optimizer='adam',  # Utilisation de l'optimiseur Adam, efficace pour la majorité des tâches d'apprentissage
    loss='categorical_crossentropy',  # Fonction de perte adaptée à la classification multi-classes (one-hot encoded)
    metrics=['accuracy']  # Suivi de la précision pour évaluer la performance du modèle
)

# Entraînement du modèle
history = model.fit(
    inputs_train, outputs_train,  # Données d'entraînement
    epochs=600,  # Nombre total d'itérations sur l'ensemble de données
    batch_size=1,  # Taille du batch, ici 1 signifie une mise à jour du modèle après chaque exemple
    validation_data=(inputs_validate, outputs_validate)  # Utilisation des données de validation pour suivre l'évolution du modèle
)
```

## Verify

Graph the models performance vs validation.


### Graph the loss

Graph the loss to see when the model stops improving.

```python
# Augmenter la taille des graphiques. La taille par défaut est (6,4).

plt.rcParams["figure.figsize"] = (20,10)

# Récupération de l'historique de perte (loss) du modèle pendant l'entraînement

loss = history.history['loss']  # Perte sur les données d'entraînement
val_loss = history.history['val_loss']  # Perte sur les données de validation
epochs = range(1, len(loss) + 1)  # Création d'une liste d'époques (de 1 à nombre d'époques total)

# Tracé de la courbe de perte

plt.plot(epochs, loss, 'g.', label='Training loss')  # Courbe de perte d'entraînement en vert
plt.plot(epochs, val_loss, 'b', label='Validation loss')  # Courbe de perte de validation en bleu
plt.title('Training and validation loss')  # Titre du graphique
plt.xlabel('Epochs')  # Axe des abscisses : nombre d'époques
plt.ylabel('Loss')  # Axe des ordonnées : valeur de la perte
plt.legend()  # Afficher la légende pour distinguer les courbes
plt.show()  # Afficher le graphique

# Afficher la taille actuelle des figures définie par plt

print(plt.rcParams["figure.figsize"])
```

### Graph the mean absolute error

[Mean absolute error](https://en.wikipedia.org/wiki/Mean_absolute_error) is another metric to judge the performance of the model.



```python
# Récupération des valeurs d'accuracy (précision) au fil des époques

accuracy = history.history['accuracy']  # Précision sur l'ensemble d'entraînement
val_accuracy = history.history['val_accuracy']  # Précision sur l'ensemble de validation

# Tracé des courbes de précision en ignorant les 100 premières époques pour une meilleure lisibilité

plt.plot(epochs[SKIP:], accuracy[SKIP:], 'g.', label='Training Accuracy')  # Précision d'entraînement (vert)
plt.plot(epochs[SKIP:], val_accuracy[SKIP:], 'b.', label='Validation Accuracy')  # Précision de validation (bleu)

# Ajout des titres et légendes pour améliorer la compréhension

plt.title('Training and Validation Accuracy')  # Titre du graphique
plt.xlabel('Epochs')  # Axe des abscisses : Nombre d'époques
plt.ylabel('Accuracy')  # Axe des ordonnées : Précision
plt.legend()  # Affichage de la légende
plt.show()  # Affichage du graphique
```

# Convert the Trained Model to Tensor Flow Lite

The next cell converts the model to TFlite format. The size in bytes of the model is also printed out.

```python
# Conversion du modèle en format TensorFlow Lite sans quantification

converter = tf.lite.TFLiteConverter.from_keras_model(model)  # Crée un convertisseur pour le modèle entraîné
tflite_model = converter.convert()  # Convertit le modèle en format TensorFlow Lite

# Sauvegarde du modèle converti sur le disque

open("vibration_model.tflite", "wb").write(tflite_model)  # Écrit le modèle converti dans un fichier .tflite

# Vérification de la taille du modèle enregistré

import os
basic_model_size = os.path.getsize("vibration_model.tflite")  # Récupère la taille du fichier en octets
print("Le modèle fait %d octets" % basic_model_size)  # Affiche la taille du modèle


```

## Encode the Model in an Arduino Header File

The next cell creates a constant byte array that contains the TFlite model. Import the generated model.h file as a tab into your Arduino sketch.

```python
import os

# Lire le fichier .tflite
tflite_model_path = "vibration_model.tflite"  # Assurez-vous que ce fichier existe
header_file_path = "model_vibration.h"

try:
    with open(tflite_model_path, "rb") as f:
        tflite_data = f.read()

    # Convertir en format hexadécimal compatible C++
    hex_data = ', '.join(f'0x{byte:02x}' for byte in tflite_data)

    # Générer le fichier .h
    with open(header_file_path, "w") as header_file:
        header_file.write("#ifndef MODEL_VIBRATION_H\n")
        header_file.write("#define MODEL_VIBRATION_H\n\n")
        header_file.write("const unsigned char model_vibration[] __attribute__((aligned(4))) = {\n")
        header_file.write(hex_data)
        header_file.write("\n};\n\n")
        header_file.write("#endif // MODEL_VIBRATION_H\n")

    # Vérifier la taille du fichier généré
    model_h_size = os.path.getsize(header_file_path)
    print(f" Fichier {header_file_path} généré avec succès ({model_h_size:,} octets).")

except FileNotFoundError:
    print(f" Erreur : Le fichier {tflite_model_path} est introuvable.")
except Exception as e:
    print(f" Une erreur est survenue : {e}")
```
