# Train Neural Network





# Parse and prepare the data

Parse the CSV files and transforms them to a format that can be used to train the fully connected neural network.

If you've recorded additional vibrations, update the `vibration` list with the names of the additional CSV files.


```python
# Importation des bibliothques ncessaires

import matplotlib.pyplot as plt  # Bibliothque pour tracer des graphiques (non utilise ici, mais utile pour l'analyse)
import numpy as np  # Bibliothque pour les calculs numriques et manipulation des matrices
import pandas as pd  # Utilise pour lire et traiter les fichiers CSV contenant les donnes
import tensorflow as tf  # Bibliothque de machine learning pour entraner le modle

# Affichage de la version de TensorFlow pour vrifier linstallation

print(f"TensorFlow version = {tf.__version__}\n")

# Dfinition des types de vibrations  classifier

VIBRATIONS = ["infinity", "HeartBeat"]  #  Liste des classes de vibrations (deux types ici)

# Dfinition des paramtres lis aux enregistrements de vibrations

SAMPLES_PER_VIBRATION = 119  #  Nombre d'chantillons par vibration
NUM_VIBRATIONS = len(VIBRATIONS)  # Nombre total de types de vibrations (2 dans ce cas)

# Cration dune matrice dencodage one-hot pour la classification

ONE_HOT_ENCODED_VIBRATIONS = np.eye(NUM_VIBRATIONS)  

# Initialisation des listes pour stocker les donnes d'entre (features) et de sortie (labels)

inputs = []  #  Liste qui contiendra les signaux de vibration (acclration et gyroscope)
outputs = []  # Liste des labels one-hot correspondant aux types de vibrations

# Parcours de chaque type de vibration pour lire et traiter les donnes correspondantes

for vibration_index in range(NUM_VIBRATIONS):
    vibration = VIBRATIONS[vibration_index]  # Slection du nom de la vibration en cours de traitement
    print(f"Processing index {vibration_index} for vibration '{vibration}'.")

    # Slection de la sortie associe dans la matrice one-hot
    
    output = ONE_HOT_ENCODED_VIBRATIONS[vibration_index]

    # Chargement des donnes du fichier CSV correspondant  la vibration actuelle
    
    df = pd.read_csv(vibration + ".csv") 

    # Suppression des valeurs manquantes pour viter les erreurs
    
    df = df.dropna().reset_index(drop=True)  

    # Calcul du nombre total denregistrements dans le fichier CSV
    
    num_recordings = int(df.shape[0] / SAMPLES_PER_VIBRATION)  
    print(f"\tThere are {num_recordings} recordings of the vibration {vibration}.")

    # Traitement des enregistrements individuels
    
    for i in range(num_recordings):
        tensor = []  # Liste temporaire pour stocker les donnes de vibration sous forme de tenseur

        # Extraction des donnes pour chaque chantillon de la vibration en cours
        
        for j in range(SAMPLES_PER_VIBRATION):
            index = i * SAMPLES_PER_VIBRATION + j  # Calcul de lindex de lchantillon actuel

            # Normalisation des donnes pour les adapter au modle (chelle entre 0 et 1)
            
            tensor += [
                (df['aX'][index] + 4) / 8,  #  Normalisation des donnes d'acclration (X)
                (df['aY'][index] + 4) / 8,  #  Normalisation des donnes d'acclration (Y)
                (df['aZ'][index] + 4) / 8,  #  Normalisation des donnes d'acclration (Z)
                (df['gX'][index] + 2000) / 4000,  # Normalisation des donnes du gyroscope (X)
                (df['gY'][index] + 2000) / 4000,  # Normalisation des donnes du gyroscope (Y)
                (df['gZ'][index] + 2000) / 4000   # Normalisation des donnes du gyroscope (Z)
            ]

        # Ajout des donnes prtraites dans la liste des entres
        
        inputs.append(tensor)

        # Ajout du label one-hot correspondant dans la liste des sorties
        
        outputs.append(output)

# Conversion des listes en tableaux numpy pour une compatibilit avec TensorFlow

inputs = np.array(inputs)  # Transformation de la liste des entres en tableau numpy
outputs = np.array(outputs)  # Transformation de la liste des labels en tableau numpy

# Indication que le traitement des donnes est termin

print("Data set parsing and preparation complete..")
```

## Randomize and split the input and output pairs for training

Randomly split input and output pairs into sets of data: 60% for training, 20% for validation, and 20% for testing.

  - the training set is used to train the model
  - the validation set is used to measure how well the model is performing during training
  - the testing set is used to test the model after training

```python
# Mlanger l'ordre des entres pour assurer une rpartition quilibre entre l'entranement, le test et la validation

num_inputs = len(inputs)
randomize = np.arange(num_inputs)
np.random.shuffle(randomize)

# Remplacer les index conscutifs (0, 1, 2, etc.) par les index mlangs

inputs = inputs[randomize]
outputs = outputs[randomize]

# Diviser les enregistrements (groupes d'chantillons) en trois ensembles : entranement, test et validation

TRAIN_SPLIT = int(0.6 * num_inputs)  # 60 % des donnes pour l'entranement
TEST_SPLIT = int(0.2 * num_inputs + TRAIN_SPLIT)  # 20 % des donnes pour le test, le reste pour la validation

inputs_train, inputs_test, inputs_validate = np.split(inputs, [TRAIN_SPLIT, TEST_SPLIT])
outputs_train, outputs_test, outputs_validate = np.split(outputs, [TRAIN_SPLIT, TEST_SPLIT])

print("Mlange et division du jeu de donnes termins.")
```

## Build & Train the Model

Build and train a [TensorFlow](https://www.tensorflow.org) model using the high-level [Keras](https://www.tensorflow.org/guide/keras) API.

```python
# Construction du modle de classification des vibrations

model = tf.keras.Sequential([  # Cration d'un modle squentiel
    tf.keras.layers.Dense(50, activation='relu'),  # Premire couche cache avec 50 neurones et activation ReLU
    tf.keras.layers.Dense(15, activation='relu'),  # Deuxime couche cache avec 15 neurones et activation ReLU
    tf.keras.layers.Dense(NUM_VIBRATIONS, activation='softmax')  # Couche de sortie avec activation Softmax pour classification
])

# Compilation du modle

model.compile(
    optimizer='adam',  # Utilisation de l'optimiseur Adam, efficace pour la majorit des tches d'apprentissage
    loss='categorical_crossentropy',  # Fonction de perte adapte  la classification multi-classes (one-hot encoded)
    metrics=['accuracy']  # Suivi de la prcision pour valuer la performance du modle
)

# Entranement du modle
history = model.fit(
    inputs_train, outputs_train,  # Donnes d'entranement
    epochs=600,  # Nombre total d'itrations sur l'ensemble de donnes
    batch_size=1,  # Taille du batch, ici 1 signifie une mise  jour du modle aprs chaque exemple
    validation_data=(inputs_validate, outputs_validate)  # Utilisation des donnes de validation pour suivre l'volution du modle
)
```

## Verify

Graph the models performance vs validation.


### Graph the loss

Graph the loss to see when the model stops improving.

```python
# Augmenter la taille des graphiques. La taille par dfaut est (6,4).

plt.rcParams["figure.figsize"] = (20,10)

# Rcupration de l'historique de perte (loss) du modle pendant l'entranement

loss = history.history['loss']  # Perte sur les donnes d'entranement
val_loss = history.history['val_loss']  # Perte sur les donnes de validation
epochs = range(1, len(loss) + 1)  # Cration d'une liste d'poques (de 1  nombre d'poques total)

# Trac de la courbe de perte

plt.plot(epochs, loss, 'g.', label='Training loss')  # Courbe de perte d'entranement en vert
plt.plot(epochs, val_loss, 'b', label='Validation loss')  # Courbe de perte de validation en bleu
plt.title('Training and validation loss')  # Titre du graphique
plt.xlabel('Epochs')  # Axe des abscisses : nombre d'poques
plt.ylabel('Loss')  # Axe des ordonnes : valeur de la perte
plt.legend()  # Afficher la lgende pour distinguer les courbes
plt.show()  # Afficher le graphique

# Afficher la taille actuelle des figures dfinie par plt

print(plt.rcParams["figure.figsize"])
```

# Convert the Trained Model to Tensor Flow Lite

The next cell converts the model to TFlite format. The size in bytes of the model is also printed out.

```python
# Conversion du modle en format TensorFlow Lite sans quantification

converter = tf.lite.TFLiteConverter.from_keras_model(model)  # Cre un convertisseur pour le modle entran
tflite_model = converter.convert()  # Convertit le modle en format TensorFlow Lite

# Sauvegarde du modle converti sur le disque

open("vibration_model.tflite", "wb").write(tflite_model)  # crit le modle converti dans un fichier .tflite

# Vrification de la taille du modle enregistr

import os
basic_model_size = os.path.getsize("vibration_model.tflite")  # Rcupre la taille du fichier en octets
print("Le modle fait %d octets" % basic_model_size)  # Affiche la taille du modle


```

## Encode the Model in an Arduino Header File

The next cell creates a constant byte array that contains the TFlite model. Import the generated model.h file as a tab into your Arduino sketch.

```python
import os

# Lire le fichier .tflite
tflite_model_path = "vibration_model.tflite"  # Assurez-vous que ce fichier existe
header_file_path = "model_vibration.h"

try:
    with open(tflite_model_path, "rb") as f:
        tflite_data = f.read()

    # Convertir en format hexadcimal compatible C++
    hex_data = ', '.join(f'0x{byte:02x}' for byte in tflite_data)

    # Gnrer le fichier .h
    with open(header_file_path, "w") as header_file:
        header_file.write("#ifndef MODEL_VIBRATION_H\n")
        header_file.write("#define MODEL_VIBRATION_H\n\n")
        header_file.write("const unsigned char model_vibration[] __attribute__((aligned(4))) = {\n")
        header_file.write(hex_data)
        header_file.write("\n};\n\n")
        header_file.write("#endif // MODEL_VIBRATION_H\n")

    # Vrifier la taille du fichier gnr
    model_h_size = os.path.getsize(header_file_path)
    print(f" Fichier {header_file_path} gnr avec succs ({model_h_size:,} octets).")

except FileNotFoundError:
    print(f" Erreur : Le fichier {tflite_model_path} est introuvable.")
except Exception as e:
    print(f" Une erreur est survenue : {e}")
```
